```python
#Hello World
#In this assignment, you’re asked to create the nodes and edges for a basic graph, such as the Krackhardt kite shown
#below.
#Below I recreated the Krackhardt kite referenced in the assignment
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()

#G.add_node("A")
G.add_nodes_from(["A","B","C","D","E","F","G","H","I","J"])
#G.add_edge(*("A","B"))
G.add_edges_from([  ("A","B"), ("A","C"), ("A","F"), ("A","J"), ("E","J")
                  , ("J","B"), ("B","D"), ("B","E"), ("C","E"), ("D","A")
                  , ("C","F"), ("C","I"), ("E","I"), ("F","G"), ("G","H")
                  , ("I", "J"), ("D", "J"), ("J","C")
                 ])
nx.draw_networkx(G)
plt.show() 
```
